```python
CRYPTO_PANIC_API_KEY = "f2b5f828c73705d7c2f8a681e5dbc53a3357af38"  # Replace with your own key
API_URL = f"https://cryptopanic.com/api/v1/posts/?auth_token={CRYPTO_PANIC_API_KEY}&currencies=BTC&public=true"
```

```python
def fetch_latest_news():
    response = requests.get(API_URL)
    if response.status_code != 200:
        print("Error fetching data:", response.status_code)
        return []

    raw_data = response.json().get("results", [])
    articles = []

    for item in raw_data:
        raw_title = item.get("title", "")
        published_at = item.get("published_at", "")
        source = item.get("domain", "")
        url = item.get("url", "")

        # Clean the title and construct a richer content string
        body = clean_text(raw_title)
        summary = f"{body}. Published on {published_at} by {source}. Full article: {url}"

        sentiment = score_sentiment(summary)

        articles.append({
            "content": summary,
            "published": published_at,
            "source": source,
            "url": url,
            "sentiment": sentiment,
            "timestamp": datetime.now(timezone.utc).isoformat()
        })

    return articles
```
