```python
# Ensure built-in types are used
from builtins import int, float
import pandas as pd
import matplotlib.pyplot as plt

# Load and prepare data
df = pd.read_csv('../data/bitcoin_cleaned.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.sort_values(by='timestamp')
df.set_index('timestamp', inplace=True)
df['price_usd'] = df['price_usd'].apply(float)

# --- Plot 1: Raw Bitcoin price ---
plt.figure(figsize=(int(14), int(5)))
plt.plot(df.index, df['price_usd'], label='Price (USD)', color='blue')
plt.title('Bitcoin Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# --- Plot 2: Raw + Rolling average ---
window_size = int(24)
df['rolling_avg'] = df['price_usd'].rolling(window=window_size).mean()

plt.figure(figsize=(int(14), int(5)))
plt.plot(df.index, df['price_usd'], alpha=0.4, label='Raw Price', color='blue')
plt.plot(df.index, df['rolling_avg'], color='red', label='24-point Rolling Average')
plt.title('Bitcoin Price with Rolling Average')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
