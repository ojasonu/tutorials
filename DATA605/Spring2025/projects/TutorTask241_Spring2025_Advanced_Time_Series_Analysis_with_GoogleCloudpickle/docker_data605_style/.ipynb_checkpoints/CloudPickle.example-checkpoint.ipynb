# CloudPickle Example - Interactive End-to-End Workflow

This notebook demonstrates the complete, interactive workflow for analyzing Bitcoin price data. It covers the following objectives:

- **Data Ingestion**: Fetch Bitcoin price data from CoinGecko using the `fetch_bitcoin_data()` function.
- **Data Serialization**: Serialize and load functions to handle moving averages and anomaly detection.
- **Time Series Analysis**: Perform time series analysis by calculating moving averages and detecting anomalies.
- **Interactivity**: Allow real-time adjustments of parameters such as the moving average window and anomaly threshold using sliders.
- **Distributed Processing**: Distribute the analysis tasks across multiple CPU cores using Python's `multiprocessing` module for improved performance.
